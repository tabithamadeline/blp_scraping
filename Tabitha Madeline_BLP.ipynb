### Multiplepage

```python
data_products = []
data_prices = []

for index in range(1,8):
  page = requests.get(f'https://blpbeauty.com/collections/makeup?page={index}/')
  soup = BeautifulSoup(page.content, 'html.parser')

  products = soup.find_all('p', class_='product--title')

  for each_product in products:
    data_products.append(each_product.text)

  prices = soup.find_all('span', class_='product--price money')

  for each_price in prices:
    data_prices.append(each_price.text)
```

```python
data = {
    'data_products' : data_products,
    'data_prices' : data_prices
}
```

```python
pd.DataFrame(data).to_csv('blp_2.csv')
```
